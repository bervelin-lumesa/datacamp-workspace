## Introduction

**Kinshasa** est la capitale et la plus grande ville de la république démocratique du Congo (RDC) ainsi que d'Afrique ; elle s’étend sur 9965 km2. Avec une population estimée à plus de 15 millions d'habitants dans sa zone métropolitaine, elle est la troisième agglomération d'Afrique derrière Le Caire et Lagos, et constitue la plus grande agglomération francophone du monde, en ayant dépassé celle de Paris dans les années 20102, et figure parmi les agglomérations les plus peuplées au monde. 

Le **[tidyverse](https://www.tidyverse.org/packages/)** est une collection de packages R conçus pour la science des données. Tous ses packages partagent une philosophie de conception, une grammaire et des structures de données sous-jacentes.

Dans cet article, nous allons découvrir Kinshasa, en utilisant les différentes fonctions issues des packages faisant partie du `tidyverse`. Il s'agira par exemple du package `dplyr` pour la manipulation de données ou `ggplot2` pour la visualisation.

```r
# Téléchargement des packages
library(tidyverse)
library(readxl) # pour lire le fichier excel
library(patchwork) # pour afficher plusieurs graphhiques dans le même panel 
```

## Données

Il s'agit d'une base de données des quartiers de la ville de Kinshasa. Elle comprend des informations sur tous les quartiers de Kinshasa, leur communes, districts et populations. Chaque ligne est correspond à un quartier.

C'est une base qui est utilisée pour le tirage des unités administratives lors des opérations de collecte de données.

```r
(kinshasa <- read_excel("kinshasa.xlsx"))
```

## Explorer

La base est constituée de 4 variables suivantes : 
1. district : Le nom du district;
2. Commune : Le nom de la commune;
3. Quartier : Le nom du quartier;
4. Population : La population du quartier en 2015.

```r
str(kinshasa)
```

Avant de continuer, actualisons la population des quartiers pour 2022. Dans la base de données, elle est celle de 2015. Nous allons utiliser le taux de croissance naturelle de 3.1% pour ce faire. Par la même occasion, nous mettons en majuscule le contenu de la variable district.

```r
kinshasa <- kinshasa %>% 
  mutate(Population = round(Population * 1.031 ^ 7),
        district = str_to_upper(district)) 
```

### Nombre de communes par district

```r
kinshasa %>% 
  group_by(district) %>% 
  summarize(n = n_distinct(Commune))
```

Les districts de FUNA et LUKUNGA ont chacun 7 communes, alors que MONT-AMBA et TSHANGU en ont 5 chacun. 

### Nombre de quartiers par commune

Selon le nombre de quartiers par commune, il ressort que KIMBANSEKE est la commune avec le plus grand nombre de quartiers, soit 44. Elle est suivie loin derrière par de la commune de NSELE avec 25 quartiers. La commune de NGABA occupe le dernier classement avec 6 quartiers.

```r
kinshasa %>% 
  count(Commune, sort = T)
```

```r
kinshasa %>% 
  count(Commune) %>% 
  ggplot(aes(x = fct_reorder(Commune, n), y = n)) +
  geom_col(fill = "cornflowerblue") + 
  coord_flip() + 
  labs(title = "Répartition des communes selon le nombre de quartiers dans la ville de kinshasa",
       subtitle = "Kimbanseke est en tête, suivi de N'sele",
       x = "Commune",
       y = "Nombre de quartiers",
       caption = "Source : Base de sondage de l'INS-RDC") + 
  theme(
    panel.grid.major.y = element_blank()
  ) + 
  geom_text(aes(label = n), hjust = 1.2, col = "white") 
```

```r
kinshasa %>% 
  group_by(Commune) %>% 
  summarise(n_quartier = n(),
            pop = sum(Population)) %>% 
  arrange(Commune) %>% 
  ggplot(aes(n_quartier, pop)) + 
  geom_point() +
  scale_y_log10() + 
  ggrepel::geom_text_repel(aes(label = Commune)) + 
  theme_classic()
```

### La population de Kinshasa

```r
kinshasa %>%  
  pull(Population) %>% 
  summary() 
  
hist <- kinshasa %>% 
  ggplot(aes(Population)) + 
  geom_histogram(binwidth = 5000)+ 
  theme_classic()

box <- kinshasa %>% 
  ggplot(aes(1, Population)) + 
  geom_boxplot() + 
  theme_classic()

hist / box
```

### Population par district

```r
kinshasa %>% 
  ggplot(aes(district, Population)) + 
  geom_boxplot() + 
  theme_classic()
```

```r
kinshasa %>%
  group_by(district) %>%
  summarise(pop = sum(Population)) %>%
  arrange(desc(pop))
```

### Commune la plus peuplée

```r
kinshasa %>%
  group_by(Commune) %>%
  summarise(pop = sum(Population)) %>%
  arrange(desc(pop))
```

```r
kinshasa %>%
  group_by(district, Commune) %>%
  summarise(pop = sum(Population)) %>%
  group_by(district) %>%
  top_n(1) %>%
  #arrange(desc(pop)) %>%
  ggplot(aes(x = Commune, y = pop, col = district)) + 
  geom_col(fill = "white") +
  theme_classic()
```

```r
kinshasa %>%
  group_by(Commune, quartier) %>%
  summarise(pop = sum(Population)) %>%
  arrange(desc(pop))
```

### Taille des quartiers 

A partir de la médiane, qui est une statistique qui dévise une distribution en deux groupes dont chacun a 50% d'observations, nous allons créer une nouvelle variable qui donne une idée sur la taille des quartiers. Cette variable aura deux modalités : Grand, si le quartier a une population supérieure à la population médiane et Petit dans le cas contraire. Cette catégorisation n'a rien de concrête, elle est juste dans un objectif pédagogique.  

```r
pop_median <- kinshasa %>% 
  summarize(median_p = median(Population)) %>% 
  pull(median_p) %>% round()

## ajouter une colonne : taille
kinshasa <- kinshasa %>% 
  mutate(taille = ifelse(Population > pop_median, "Grand", "Petit"))
```

```r
(pl1 <- kinshasa %>% 
  ggplot(aes(Commune, fill = taille)) + 
  geom_bar(position = "fill") + 
  coord_flip() + 
  labs(fill = "Taille des quartiers") + 
  scale_y_continuous("%", labels = paste0(seq(0, 100, 25))) + 
  theme(legend.position = "bottom"))
```

### Taille des quartiers dans chaque district

```r
pl1 + 
  facet_wrap(~ district, scales = "free_y") 
```
